### 1. Compute the annual geometric returns and standard deviations for:


(a) the equity total return index,re,t.


```python
df = pd.read_excel('uk.xlsx')
```

```python
df.head()
#df.dropna(inplace=True)
```

```python
len(df)
```

```python
df['real_equity_ret']=df['uk_equity_return']-df['uk_cpi']
```

```python
df.head()
```

```python
df['real_gross_equity_ret']=df['real_equity_ret']+1
```

```python
rgequity = (scipy.stats.mstats.gmean(df['real_gross_equity_ret']))**12-1
print("Annual Percent Geo. Mean of Equity Return = {:,.4f}".format(rgequity*100))
```

```python
df['log_real_gross_equity_ret']=np.log(df['real_gross_equity_ret'])
sdg_equity_ret=df['log_real_gross_equity_ret'].std()*(12**0.5)
print("Annual Percent Geo. Std. Dev. of Equity Return = {:,.4f}".format(sdg_equity_ret*100))
```

(b) the Treasury bill total return index, rf,t.

```python
df['real_bill_ret']=df['uk_bill_yield']-df['uk_cpi']
df.head()
```

```python
df['real_gross_bill_ret']=df['real_bill_ret']+1
```

```python
rgbond = (scipy.stats.mstats.gmean(df['real_gross_bill_ret']))**12-1
print("Annual Percent Geo. Mean of Bill Return = {:,.4f}".format(rgbond*100))
```

```python
df['log_real_gross_bill_ret']=np.log(df['real_gross_bill_ret'])
sdg_bill_ret=df['log_real_gross_bill_ret'].std()*(12**0.5)
print("Annual Percent Geo. Std. Dev. of Bill Return = {:,.4f}".format(sdg_bill_ret*100))
```

(c) the equity premium, defned as the equity return less the Treasury bill return, re,tâˆ’rf,t.

```python
df['real_equity_premium']=df['real_equity_ret']-df['real_bill_ret']
df.dropna(inplace=True)
```

```python
rpremium = (scipy.stats.mstats.gmean(df['real_equity_premium']+1))**12-1
print("Annual Percent Equity Premium = {:,.4f}".format(rpremium*100))
```

```python
df['real_gross_equity_premium']=df['real_equity_premium']+1
df['log_real_gross_equity_premium']=np.log(df['real_gross_equity_premium'])
# sdg_equity_premium=math.sqrt((scipy.stats.moment(crsp_data['log_real_gross_equity_premium'],moment=2)))*math.sqrt(12)
sdg_equity_premium=df['log_real_gross_equity_premium'].std()*(12**0.5)
print("Annual Percent Geo. Std. Dev. of Equity Premium = {:,.4f}".format(sdg_equity_premium*100))
```

(d) private final consumption expenditures ct.

```python

df['real_cons_growth']=np.log(df['uk_cons'])-np.log(df['uk_cons'].shift(1))
df.dropna(inplace=True)
df.head()
```

```python
df['real_gross_cons_growth']=df['real_cons_growth']+1
```

```python
rgcons = (scipy.stats.mstats.gmean(df['real_gross_cons_growth']))**12-1
print("Annual Percent Geo. Mean of Cons. Growth = {:,.4f}".format(rgcons*100))
```

```python
df['log_real_gross_cons_growth']=np.log(df['real_gross_cons_growth'])
sdg_real_cons_growth=df['log_real_gross_cons_growth'].std()*(12**0.5)
print("Annual Percent Geo. Std. Dev. of Cons. Growth = {:,.4f}".format(sdg_real_cons_growth*100))
```
